Matplotlib 的API函数都位于 matplotlib.pyplot模块中

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

### 一个简单的范例

```python
import numpy as np
x = np.linspace(-1,1,50)
y = 2*x+1
plt.plot(x,y)#第一个是横坐标的值，第二个是纵坐标的值
plt.show()
```

## 显示多个图像

- 使用figure()函数重新申请一个figure对象
- 每次调用figure的时候都会重新申请一个figure对象
- figure的num表示其编号，figsize表示的是图表的长宽
- 当我们需要在画板中绘制两条线的时候，可以开一个figure后，然后用两个plot

```python
x = np.linspace(-1,1,50)
y1 = 2*x + 1
y2 = 2**x + 1

plt.figure()
plt.plot(x,y1)
plt.figure(num = 3,figsize = (8,5))
plt.plot(x,y2)
plt.plot(x,y1,color = 'red',#线颜色
         linewidth = 1.0,#线宽
         linestyle = '--'#线样式
        )
plt.show()
```

### 去除边框，指定轴的名称

```python
plt.figure()
plt.plot(x,y1)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

### 同时绘制多条曲线

```python
plt.figure(num = 3,figsize = (8,5))
# num表示的是编号，figsize表示的是图表的长宽
plt.plot(x,y2)
plt.plot(x,y1,
        color = 'red',
        linewidth = 1.0,
        linestyle = '--')
#下面两句表示的是画图的区域范围
plt.xlim((-1,2))
plt.ylim((1,3))

new_ticks = np.linspace(-1,2,5)#-1到2等步长取5个值
plt.xticks(new_ticks)#将这五个值等量放入这里面
plt.yticks([-2, -1.8, -1, 1.22, 3],
          ['really bad', r'$really\ bad$', r'$normal$', r'$good$', r'$readly\ good$'])
#这里面不一样的地方在于字体方面可能会有些区别，为点的位置设置对应的文字，每个参数是点的位置，第二参数是点的文字提示
#gca即get current axis
ax = plt.gca()
#将右边和上边的边框的颜色去掉
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#绑定x轴和y轴
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
#定义x轴和y轴的位置
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

plt.show()
```

### 多条曲线之曲线说明

这里面有非常重要的地方是要加一个逗号，然后那个显示label才不会报错

```python
plt.figure(num = 3,figsize = (8,5))
plt.plot(x,y2)
plt.plot(x,y1,color = 'red', linewidth = 1.0, linestyle = '--')
plt.xlim((-1,2))
plt.ylim((1,3))
plt.xlabel("I am x")
plt.ylabel("I am y")
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)
plt.yticks([-2, -1.8, -1, 1.22,3],
          [r'$really\ bad$', r'$bad$', r'$normal$', r'$good$', r'$readly\ good$'])

l1, = plt.plot(x, y2, 
               label='aaa'
              )#l1后面的逗号非常重要，不然不会出现label
l2, = plt.plot(x, y1, 
               color='red',  # 线条颜色
               linewidth = 1.0,  # 线条宽度
               linestyle='-.',  # 线条样式
               label='bbb'  #标签
              )

#使用legend绘制多条曲线
plt.legend(handles = [l1,l2],
          labels = ['aaa','bbb'],
          loc = 'best')



plt.show()
```

### 多个figure,并加上特殊点的注释

```python
plt.figure(figsize=(12, 8))
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

#显示交叉点
x0 = 1
y0 = 2*x0 +1

#scatter用来显示这个点，s表示的是点的大小
plt.scatter(x0,y0, s=50,color = 'r')

#设置线的范围，x的范围是定值，y的范围是从y0到0的位置
#lw的意思是linewidth,线宽
plt.plot([x0,x0],[y0,0],'k-.',lw= 2.5)

#设置关键位置的提示信息
plt.annotate(r'$2x+1=%s$'%y0,#提示的信息
            xy = (x0,y0),#提示的位置
            xycoords = 'data',
            xytext = (+30,-30),#这个地方应该是在设置提示信息与关键点的位置关系
            textcoords = 'offset points',
            fontsize = 16,#这里设置的是字体的大小
             #这里设置的是箭头和箭头的弧度
            arrowprops = dict(arrowstyle = '->',connectionstyle = 'arc3,rad=.2')
            )

plt.text(0, 3, 
         r'$This\ is\ a\ good\ idea.\ \mu\ \sigma_i\ \alpha_t$',
         fontdict={'size':16,'color':'r'})
plt.show()
```

### tick能见度设置

```python
# 从[-1,1]中等距去50个数作为x的取值
x = np.linspace(-1, 1, 50)
y = 2*x - 1

plt.figure(figsize=(12, 8))  # 第一个参数表示的是编号，第二个表示的是图表的长宽
# alpha是设置透明度的
plt.plot(x, y, color='r', linewidth=0.6, alpha=0.5)

# gca = 'get current axis'
ax = plt.gca()
# 将右边和上边的边框（脊）的颜色去掉
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# 绑定x轴和y轴
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# 定义x轴和y轴的位置
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

# 可以使用tick设置透明度
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(12)
    label.set_bbox(dict(facecolor='y', edgecolor='None', alpha=0.7))

plt.show()
```

### 条形图

```python
n = 12
X = np.arange(n)
Y1 = (1 - X/float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X/float(n)) * np.random.uniform(0.5, 1.0, n)

plt.figure(figsize=(12, 8))
#设置了一个bar来做条形图
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x, y in zip(X,Y1):
    # ha: horizontal alignment水平方向
    # va: vertical alignment垂直方向
    plt.text(x, y+0.05, '%.2f' % y, ha='center', va='bottom')
    #这个0.05就是距离条形图y的距离

for x, y in zip(X,-Y2):
    # ha: horizontal alignment水平方向
    # va: vertical alignment垂直方向
    plt.text(x, y-0.05, '%.2f' % y, ha='center', va='top')

# 定义范围和标签
plt.xlim(-.5, n)
plt.xticks(())#将坐标轴上的值全部删去
plt.ylim(-1.25, 1.25)
plt.yticks(())

plt.show()
```

### contour等高线图

```python
def get_height(x, y):
    # the height function
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)

plt.figure(figsize=(14, 8))

# use plt.contourf to filling contours
# X, Y and value for (X, Y) point

# 横坐标、纵坐标、高度、 、透明度、cmap是颜色对应表
# 等高线的填充颜色
plt.contourf(X, Y, get_height(X, Y), 16, alpah=0.7, cmap=plt.cm.hot)  

# use plt.contour to add contour lines
# 这里是等高线的线
C = plt.contour(X, Y, get_height(X, Y), 16, color='black', linewidth=.5)

# adding label
plt.clabel(C, inline=True, fontsize=16)

plt.xticks(())
plt.yticks(())
plt.show()
```

### image图片显示

```python
# image data
a = np.array([0.313660827978, 0.365348418405, 0.423733120134,
              0.365348418405, 0.439599930621, 0.525083754405,
              0.423733120134, 0.525083754405, 0.651536351379]).reshape(3,3)

"""
for the value of "interpolation", check this:
http://matplotlib.org/examples/images_contours_and_fields/interpolation_methods.html
for the value of "origin"= ['upper', 'lower'], check this:
http://matplotlib.org/examples/pylab_examples/image_origin.html
"""

# 这是颜色的标注
# 主要使用imshow来显示图片，这里暂时不适用图片来显示，采用色块的方式演示。
plt.imshow(a, interpolation='nearest', cmap='bone', origin='lower')
plt.colorbar(shrink=.90)  # 这是颜色深度的标注，shrink表示压缩比例

plt.xticks(())
plt.yticks(())
plt.show()
```

### 3D数据图

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 8))
ax = Axes3D(fig)
#将这张fig标志成3D图

# 生成X，Y
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X,Y = np.meshgrid(X, Y)#将两个维度进行一个合成，需要用两个变量来接收
R = np.sqrt(X**2 + Y**2)

# height value
Z = np.sin(R)

# 绘图
# rstride（row）和cstride(column)表示的是行列的跨度
ax.plot_surface(X, Y, Z, 
                rstride=1,  # 行的跨度
                cstride=1,  # 列的跨度
                cmap=plt.get_cmap('rainbow')  # 颜色映射样式设置
               )

# offset 表示距离zdir的轴距离
ax.contourf(X,Y,Z, zdir='z', offest=-2, cmap='rainbow')
ax.set_zlim(-2, 2)

plt.show()
```

### 分格显示（subplot2grid）

```python
import matplotlib.gridspec as gridspec
plt.figure()

ax1 = plt.subplot2grid((3,3),(0,0),colspan = 3,rowspan =1)
ax1.plot([1,2],[1,2])#第一个元素表示的是x的范围是[1,2],后面一个表示的是y的范围是[1,2]
ax1.set_title('ax1_title')
#第一个元素表示将总的面板进行划分，划分为3行3列
#第二个元素表示该面板从1行0列开始，列的跨度（colspan）为2列，行的跨度为默认值
ax2 = plt.subplot2grid((3,3),(1,0),colspan = 2)
ax2.set_title('ax2_tilte')

plt.tight_layout()#是否要密致布局
plt.show()
```

### 网格布局（gridspec）

```python
plt.figure()
gs = gridspec.GridSpec(3,3)
ax1 = plt.subplot(gs[0,:])#这里表示从0行开始全部都是ax1的
ax1.set_title('ax1_title')

ax2 = plt.subplot(gs[1,:2])
ax2.set_title('ax2_tltle')

ax3 = plt.subplot(gs[1:,2])
ax3.set_title('ax3_title')

ax4 = plt.subplot(gs[-1,0])
ax4.set_title('ax4_title')

ax5 = plt.subplot(gs[-1,-2])
ax5.set_title('ax5_title')

plt.tight_layout()
plt.show()
```

### 分格显示（easy to define structure）

```python
plt.figure()
#sharex 表示共享x轴，sharey表示共享y轴
f,((ax11,ax12),(ax21,ax22)) = plt.subplots(2,2,sharex = True, sharey = True)

ax11.scatter([1,2],[1,2])
ax11.set_title('ax11_title')
ax12.set_title('ax12_title')
ax21.set_title('ax21_title')
ax22.set_title('ax22_title')

plt.tight_layout()
plt.show()
```

### 图中图

```python
fig = plt.figure(figsize = (10,6))
x = [1,2,3,4,5,6,7]
y = [1,3,4,2,5,8,6]

#大图
left, bottom, width, weight = 0.1,0.1,0.8,0.8
#add_axes是对于绘图布局进行一个设置
ax1 = fig.add_axes([left,bottom,width,weight])
ax1.plot(x,y,'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('ax1_title')

#左上小图
left,bottom,width,weight = 0.2,0.6,0.25,0.25
ax2 = fig.add_axes([left,bottom,width,weight])
ax2.plot(x,y,'b')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('title_inside')

#右下小图
plt.axes([0.6,0.2,0.25,0.25])#这个每个数都是一个比例值
plt.plot(x,y[::-1],'g')#将y进行逆序输出
plt.xlabel('x')
plt.ylabel('y')
plt.title('title_inside')

plt.show()
```
